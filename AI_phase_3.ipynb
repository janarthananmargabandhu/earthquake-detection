## Parsing dates
when we are using a timeseries data, we want to enrich the time and date component as much as possible.

we can do it buy telling pandas which of the colomns has dates in it using the parameter parse_dates.

```python
df = pd.read_csv("csv files\database.csv",
                 low_memory=False,
                 parse_dates=["Date"])
```


## Convert string into numbers

One way to convert all the data into numbers is by converting them into pandas categories.

```python
pd.api.types.is_string_dtype(df.Source)
```

```python
# Find the columns which contains string
for label, content in df.items():
    if pd.api.types.is_string_dtype(content):
        print(label)
```

```python
# Converting all string values into categories
for label, content in df.items():
    if pd.api.types.is_string_dtype(content):
        df[label] = content.astype("category").cat.as_ordered()
```
